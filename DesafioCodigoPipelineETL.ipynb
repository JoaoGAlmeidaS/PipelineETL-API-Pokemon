# Extract & Transform

```python
import pandas as pd
import requests
import json

#pegar apenas os pokemons da 1 geração
url = "https://pokeapi.co/api/v2/pokemon?limit=151"
```

```python
def get_pokemon_data():
  response = requests.get(url)

  #200 é o status code de quando o request da certo
  if response.status_code == 200:
    return response.json()
  else:
    print("Erro ao obter os resultados")
    return None

def extract_poke_info(data):
  poke_info = []

  for pokemon in data['results']:
    poke_details = requests.get(pokemon['url']).json()
    name = poke_details['name']
    dex_number = poke_details['id']
    types = [t['type']['name'] for t in poke_details['types']]

    if len(types) == 2:
      poke_info.append({
        'nome': name,
        'dex_number': dex_number,
        'types': types
      })

  return poke_info
```

```python
poke_data = get_pokemon_data()

if poke_data:
  pokemons = extract_poke_info(poke_data)
  print(f"Pokémons com 2 tipos extraídos: {len(pokemons)}")
  for pokemon in pokemons:
    print(f"Nome: {pokemon['nome'].capitalize()} | Número: {pokemon['dex_number']} | Tipo: {', '.join(pokemon['types'])}\n")
```

# Load

```python
import csv

def save_to_csv(pokemons, filename="pokemons.csv"):
    keys = ['nome', 'dex_number', 'types']
    with open(filename, mode='w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=keys)
        writer.writeheader()
        for pokemon in pokemons:
            # Transformando o campo 'types' em string
            pokemon['types'] = ', '.join(pokemon['types'])
            writer.writerow(pokemon)

# Salvando os dados em um CSV
save_to_csv(pokemons)
```
